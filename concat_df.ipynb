```python
import pandas as pd

df_main = pd.read_csv("./game_data.csv")
df_keywords = pd.read_csv("./game_keywords.csv")
```

```python
concat_key = []
```

```python
df_keywords
for i in range(len(df_keywords)):
    a = df_keywords['game_name'][i].split("\\")[1]
    concat_key.append(a.split(".csv")[0])
```

```python
len(concat_key)
```

```python
concat_keywords = []

for i in range(len(concat_key)):
    for j in range(len(df_main['url'])):
        if concat_key[i] in df_main['url'][j]:
            concat_keywords.append([df_keywords['keywords'][i] , j])
            break
```

```python
concat_keywords.sort(key= lambda x: x[1])
```

```python
len(concat_keywords)
```

```python
final = []
for c in concat_keywords:
    final.append(c[0])
    
final
```

```python
df_main['key_words'] = final
```

```python
df_main
```

```python
df_main.to_csv('game_info_keywords.csv', index=False)
```
